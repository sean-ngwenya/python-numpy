```python
import numpy as np

my_matrix = np.random.randint(1, 101, (5, 5))

my_mean = np.mean(my_matrix)

my_std = np.std(my_matrix)

# set elements greater than 50 to 0

greater_50_elements = my_matrix[my_matrix > 50]
my_matrix[my_matrix > 50] = 0

print("Modified Matrix:\n", my_matrix)

print("Elements greater than 50 set to 0:\n", greater_50_elements)
print("Mean:", my_mean)
print("Standard Deviation:", my_std)    
```
